```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


barajlar = ["BalÃ§ova BarajÄ±","GÃ¶rdes BarajÄ±","TahtalÄ± BarajÄ±","Ãrkmez BarajÄ±","GÃ¼zelhisar BarajÄ±","AlaÃ§atÄ± Kutlu AktaÅ BarajÄ±"]
doluluk_oranlari = [12.96, 1.57,	12.05,	6.55,	64.98,	7.54]


data = pd.DataFrame({
    "Baraj": barajlar,
    "Doluluk OranÄ± (%)": doluluk_oranlari
})

plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")


sns.barplot(x="Doluluk OranÄ± (%)", y="Baraj", data=data, palette="coolwarm")

plt.title("Ä°zmir BarajlarÄ± Doluluk OranlarÄ±", fontsize=16)
plt.xlabel("Doluluk OranÄ± (%)", fontsize=12)
plt.ylabel("Barajlar", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(doluluk_oranlari, labels=barajlar, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))

plt.title("Ä°zmir BarajlarÄ± Doluluk OranlarÄ± (Pasta GrafiÄi)", fontsize=16)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(barajlar, doluluk_oranlari, marker='o', linestyle='-', color='b', label="Doluluk OranÄ±")

plt.title("Ä°zmir BarajlarÄ± Doluluk OranlarÄ±", fontsize=16)
plt.xlabel("Barajlar", fontsize=12)
plt.ylabel("Doluluk OranÄ± (%)", fontsize=12)
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```
